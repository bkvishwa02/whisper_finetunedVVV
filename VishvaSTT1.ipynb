```python
def download_video_mp4(youtube_url):
  # Create a YouTube object
  yt = YouTube(youtube_url)

  # Get the video with the highest resolution and file size
  video = yt.streams.filter(progressive=True,
                            file_extension='mp4').order_by('resolution').asc().first()
  # Download the video to the current working directory
  video.download()

  print('Video downloaded!')
  return 1
```

```python
import librosa
import datasets
from transformers import WhisperProcessor
import numpy as np  # For potential filtering logic

# --- Dataset and Processor Setup ---
dataset = datasets.load_dataset("mozilla-foundation/common_voice_11_0", "en", use_auth_token=True, streaming=True)
processor = WhisperProcessor.from_pretrained("openai/whisper-medium")

# --- Preprocessing Function ---
def preprocess(example):
    # Load and resample audio
    audio_array = example["audio"]["array"]
    audio_array = librosa.resample(audio_array, example["audio"]["sampling_rate"], 16000)

    # Normalize audio
    audio_array = librosa.util.normalize(audio_array)

    # Lowercase text
    text = example["sentence"].lower()

    # Prepare for Whisper
    input_features = processor(audio_array, sampling_rate=16000, return_tensors="pt")
    labels = processor(text, return_tensors="pt").input_ids

    return {"input_values": input_features.input_values, "labels": labels}

# --- Apply Preprocessing and Training Prep ---
processed_dataset = dataset["train"].map(preprocess, batched=True, batch_size=4, remove_columns=dataset["train"].column_names)
```
